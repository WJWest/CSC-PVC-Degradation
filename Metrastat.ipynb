### Vary k12 (rate constant for removal of HCL) to see effect on concentration profile

```python
from __future__ import division
from numpy import linspace
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from parameters import get_constants, get_initial
from Metrastat import metrastat
% pylab inline

# initial conditions
initial = get_initial()

# plotting information
timespan = linspace(0, 120, 120)
legendstr = ['HCl',
             'LDH',
             'Polymer active sites',
             'Radical',
             'Primary stabiliser',
             'Degraded polymer',
             'Cross-links']

def kfit(k12_vary):

    
    simulation = odeint(metrastat, initial, timespan, args=(k12_vary,))
    
    plt.plot(timespan, simulation)
    plt.title('Specie concentrations')
    plt.xlabel('Time /min')
    plt.ylabel('Concentration')
    plt.legend(legendstr)

from IPython.html.widgets import interact
interact(kfit, k12_vary=(0.1, 10, 0.1))
```
